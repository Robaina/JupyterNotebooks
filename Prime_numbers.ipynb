## Counting twin primes in intervals

```python
from sympy.ntheory.generate import primerange
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
#plt.style.use('dark_background')
%matplotlib inline
    
def count_twin_primes(a, b):
    """
    Counts twin primes in interval [a, b]
    """
    primes = primerange(a, b)
    count = 0
#     twins = []
    for prime in primes:
        try:
            next_prime = next(primes)
            if next_prime - prime == 2:
                count += 1
#                 twins.append((prime, next_prime))
        except StopIteration:
            None
    return count
```

```python
time_period = 10000
N = 1000
years = [n * time_period for n in range(N)]
number_of_twins = []
for i in range(0, N):
    a, b = i * time_period, (i + 1) * time_period
    number_of_twins.append(count_twin_primes(a, b))
    
plt.figure(figsize=(12, 9))
plt.title('Twin primes per ' + str(time_period) + ' years', fontsize=14)
plt.xlabel('year', fontsize=14)
plt.ylabel('counts', fontsize=14)
plt.scatter(years, number_of_twins)
plt.show()
```
