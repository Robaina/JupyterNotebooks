# A trascendental equation
A [trascendental equation](https://en.wikipedia.org/wiki/Transcendental_equation) generally does not have closed-form solutions. An example of such an equation is:
$$x^y = y^x,$$
which has the trivial solution $x = y$, but also a non-trivial solution that cannot be expressed in a [closed form](https://en.wikipedia.org/wiki/Closed-form_expression). Instead, we will plot all points $(x, y) \in \mathrm{I\!R}^2$ satisfying the equation above to visualize the solution.

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
dx = dy = 0.01
xmax = ymax = 12
x, y = np.arange(0, xmax, dx), np.arange(0, ymax, dy)
X, Y = np.meshgrid(x, y)
F = X**Y - Y**X

plt.figure(figsize=(8, 8))
plt.title('Solutions to $x^y=y^x$', fontsize=18)
plt.xlabel('x', fontsize=18)
plt.ylabel('y', fontsize=18)
plt.contour(X, Y, F, [0], colors='blue')
plt.hlines(np.e, 0, np.e, linestyles='dashed')
plt.vlines(np.e, 0, np.e, linestyles='dashed')
plt.text(np.e, -.5, '$e$', fontsize=18)
plt.text(-.5, np.e, '$e$', fontsize=18)
plt.text(4.1, 8.5, '$\mathrm{I}\;(x^y > y^x)$', fontsize=14)
plt.text(8.5, 5, '$\mathrm{II}\;(x^y < y^x)$', fontsize=14)
plt.text(6, 0.5, '$\mathrm{III}\;(x^y > y^x)$', fontsize=14)
plt.text(0.05, 3.3, '$\mathrm{IV}\;(x^y < y^x)$', fontsize=14)
plt.show()
```

As we see above, the intersection of the two sets of solutions occurs at $(e, e)$! Moreover, $y \rightarrow 1$ when $x \rightarrow \infty$ and vice versa. The two curves divide the plane in four regions, in which either of the inequalities $x^y > y^x$, $x^y < y^x$ hold, which we can find out by testing some points within the regions. Interestingly, when both numbers in any pair are greater than $e$, _i.e._ we are in regions I and II, the following rule emerges:

$$x < y \rightarrow x^y > y^x$$
$$x > y \rightarrow x^y < y^x,$$

for instance, setting $x = e, y = \pi$, since $e < \pi$ we have $e^{\pi} > \pi^e$.
